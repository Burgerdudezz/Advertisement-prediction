# Quick EDA – Internet Advertisements Dataset

This notebook performs a lightweight exploratory data analysis on the **cleaned** dataset.
Run `python -m src.run_pipeline` (at least the preprocess step) before opening this notebook.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.config import PROCESSED_DATA_PATH, TARGET_COLUMN

sns.set_theme(style='whitegrid')
```

## 1  Load the cleaned data

```python
df = pd.read_csv(PROCESSED_DATA_PATH)
print(f'Shape: {df.shape}')
df.head()
```

## 2  Basic info & missing values

```python
df.info()
print('\nMissing values per column:')
print(df.isnull().sum().sort_values(ascending=False).head(10))
```

## 3  Class distribution

```python
class_counts = df[TARGET_COLUMN].value_counts()
print(class_counts)

fig, ax = plt.subplots(figsize=(5, 3))
class_counts.plot(kind='bar', ax=ax, color=['steelblue', 'salmon'])
ax.set_title('Class distribution')
ax.set_xlabel('Class')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

## 4  Summary statistics for numeric features

```python
df.describe().T.sort_values('std', ascending=False).head(20)
```

## 5  Correlation heat-map (first 20 numeric features)

```python
numeric_df = df.select_dtypes(include='number').iloc[:, :20]

fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm',
    center=0,
    linewidths=0.4,
    ax=ax,
)
ax.set_title('Correlation matrix (first 20 numeric features)')
plt.tight_layout()
plt.show()
```
